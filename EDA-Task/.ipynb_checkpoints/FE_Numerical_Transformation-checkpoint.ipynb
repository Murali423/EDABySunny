#### Transforming variables with the logarithm


The logarithm function is commonly used to transform variables. It has a strong effect on
the shape of the variable distribution and can only be applied to positive variables. In this
recipe, we will learn how to perform logarithmic transformation using NumPy, scikit-learn,
and Feature-engine. We will also create a diagnostic plot function to evaluate the effect of
the transformation on the variable distribution.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from  sklearn.datasets import load_boston
from  sklearn.preprocessing import FunctionTransformer
%matplotlib inline
#from feature_engine.variable_transformers import LogTransformer
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
boston_data = load_boston()
data = pd.DataFrame(boston_data.data,columns=boston_data.feature_names)
data
```

```python
def diagnostic_plots(df, variable):
    plt.figure(figsize=(15,6))
    plt.subplot(1, 2, 1)
    df[variable].hist(bins=30)
    plt.subplot(1, 2, 2)
    stats.probplot(df[variable], dist="norm", plot=plt)
    plt.show()
```

```python
diagnostic_plots(data, 'LSTAT')
```

The following output shows that LSTAT is not normally distributed

```python
data_df = data.copy()
data_df
```

```python
data_df[['LSTAT', 'NOX', 'DIS', 'RM']] = np.log(data[['LSTAT','NOX', 'DIS', 'RM']])
```

```python
diagnostic_plots(data_df, 'LSTAT')
```

```python
#using the sklearn 
transformer = FunctionTransformer(np.log)
```

```python
dsk = transformer.transform(data[['LSTAT','NOX', 'DIS', 'RM']])
```

```python
dsk
```

### Using square and cube root to transform variables
The square and cube root transformations are two specific forms of power transformations
where the exponents are 1/2 and 1/3, respectively. In this recipe, we will implement square
and cube root transformations using NumPy and scikit-learn.
The square root transformation is not defined for negative values, so make
sure you only transform those variables whose values are >=0; otherwise,
you will introduce NaN or receive an error message.


```python
data_tf2 = data.copy()
data_tf2
```

```python
data_tf2[['LSTAT', 'NOX', 'DIS', 'RM']] = np.sqrt(data[['LSTAT','NOX', 'DIS', 'RM']])
#data_tf[['LSTAT', 'NOX', 'DIS', 'RM']] = np.cbrt(data[['LSTAT','NOX', 'DIS', 'RM']])

data_tf2
```

```python
#using sklearn
transformer = FunctionTransformer(np.sqrt)
dsqrt = transformer.transform(data[['LSTAT', 'NOX', 'DIS','RM']])
dsqrt
```

### Using power transformations on numerical variables
Exponential or power functions are mathematical transformations that follow ,
where lambda can be any exponent. The square and cube root transformations are special
cases of power transformations where lambda is 1/2 or 1/3, respectively. In practice, we try
different lambdas to determine which one offers the best transformation. In this recipe, we
will carry out power transformations using NumPy, scikit-learn, and Feature-engine.

```python
#from feature_engine.variable_transformers import PowerTransformer
```

```python
data_tf4 = data.copy()
```

```python
data_tf4[['LSTAT', 'NOX', 'DIS', 'RM']] = np.power(data[['LSTAT','NOX', 'DIS', 'RM']],0.3)
data_tf4
```

```python
#Using sklearn
transformer = FunctionTransformer(lambda x:np.power(x,0.3))
dpow = transformer.transform(data[['LSTAT', 'NOX', 'DIS','RM']])
dpow
```

### Performing Box-Cox transformation on numerical variables
The Box-Cox transformation belongs to the power family of functions and is defined by
 if X > 0 or log(X) if X = 0, where X is the variable and Î» is the transformation
parameter. In the Box-Cox transformation, several values of Î» are considered and the Î» that
returns the best transformation is selected. In this recipe, we will perform Box-Cox
transformation using SciPy, scikit-learn, and Feature-engine.
The Box-Cox transformation can only be used on positive variables. If
your variables have negative values, try the Yeo-Johnson transformation

```python
data_tf5 = data.copy()
```

```python
data_tf5['LSTAT'], param = stats.boxcox(data['LSTAT'])  # scipy.stats.boxcox() can only be applied to one-dimensional data,and returns two parameters: the transformed variable and the optimallambda for the transformation
```

```python
param
```

```python
data_tf5['LSTAT']
```

```python
#using sklearn
from sklearn.preprocessing import PowerTransformer
```

```python
transformer = PowerTransformer(method='box-cox', standardize=False)
```

```python
cols = ['LSTAT', 'NOX', 'DIS', 'RM']
transformer.fit(data[cols])
```

```python
dbc = transformer.transform(data[cols])
dbc
```

### Performing Yeo-Johnson transformation on numerical variables

The Yeo-Johnson transformation is an extension of the Box-Cox transformation and can be
used on variables with zero and negative values, as well as positive values. 

```python
data_tf8 = data.copy()
```

```python
data_tf8['LSTAT'], param = stats.yeojohnson(data['LSTAT'])
param
```

```python
data_tf8['LSTAT']
```

```python
# using sklearn
transformer = PowerTransformer(method='yeo-johnson')
cols = ['LSTAT', 'NOX', 'DIS', 'RM']
transformer.fit(data[cols])
dyj = transformer.transform(data[cols])
dyj
```
